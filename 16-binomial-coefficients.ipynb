### 16.1 
Mixed Matched Marvin has a drawer full of 30 different socks (no two are the same).  He reaches in and grabs two.  In how many different ways can he do this?  Now he puts them on his feet (presumably, one on the left and the other on the right).  In how many different ways can he or she do that? 

#### Answer
We can map each sock from its unique pattern to the numbers $\{1, 2, \dots, 30 \}$. Grabbing two different socks is akin to creating a subset of size 2.  We therefore have ${30 \choose 2}$ different ways of doing this.  

The second question can be interpreted as how many different size 2 *lists* are possible.  We can simply use the permutation formula which is $\frac{30!}{(30-2)!}$.  This can be interpreted as counting all sub-sets of size 2 which do not care about order.  Since each subset is size 2, then we need to also count the sub-set that has the opposite order. So, we just ${30 \choose 2}$ by 2 to get $2\cdot {30 \choose 2} = \frac{2\cdot30!}{2! 28!} = \frac{30!}{28!}$

### 16.3
* How many $n$-digit binary $(0, 1)$ sequences contains exactly $k$ 1s? 
#### Answer
If we have a binary string of length $n$, then that means $k$ characters are 1's and $n-k$ characters are 0's.  This is similar to the problem of expanding $(x+y)^n$ where we count the number of times an $x$ occurs and converting its appearance to its index.  This comes down to a simple subset counting problem.  We have: 
$$
n \choose k
$$

* How many $n$-digit binary $(0, 1, 2)$ sequences contains exactly $k$ 1s? 
#### Answer
We can simply apply the same logic above on this problem and map the index of when a 1 occurs to its index.  This is akin to finding a subset of size $k$ of $\{1, \dots, n \}$.  The tricky part of this problem is that the $n-k$ other elements are not necessarily 0's.  Each of those $n-k$ elements can be either a $0$ or a $2$, so that can be counted using the multiplication principle: $ 2^{n-k}$.  We therefore have:
$$
{n \choose k} 2^{n-k}
$$

We have three actions. We know that in a single action we can press numbers one to five once, so this gives us 5 options.  We could also press two of them simultaneously and we have ${5 \choose 2} = 10$ options here.  So in a single action we have $10+5=15$ options.  Because repetition of an option is allowed, our answer is 
$$
(10+5) \cdot (10+5) \cdot (10+5) = 15^3
$$
--- 

* How many combinations are possible if no digit is repeated in the combination? 
#### Answer 
    * Case 1: Three single button inputs.
    This is $5 \cdot 4 \cdot 3$

    * Case 2: Two Single button inputs and one pair of buttons pressed simulatenously.  We need to decide when to press a pair of buttons simultaneously and since we have three actions then this is $3\choose 1$.  The remaining two actions are single button inputs and have 5 and 4 options respectively.  The pair of button pressed simultaneously action happens with $3 \choose 2$.  Putting this together we have, ${3 \choose 1 }5 \cdot 4 \cdot {3 \choose 2}$. 
    
    * Case 3: One Single button input and two pairs of buttons pressed simulatenously

    Adding all of these cases looks like:

5 \cdot 4 \cdot 3 + {3 \choose 1 }5 \cdot 4 \cdot {3 \choose 2} + {3 \choose 1} 5 \cdot {4 \choose 2} {2 \choose 2} = 60 + 3 \cdot 5 \cdot 4 \cdot 3 + 3 \cdot 5 \cdot 3! \cdot 1 = 60 + 180 + 90 = 330

### 16.8
Look down the middle column of Pascal's triangle. Notice that except for the very top 1, all these numbers are even. Why? 

#### Answer 
The middle column elements are $2={2 \choose 1}, 6 = {4 \choose 2}, 20={6 \choose 3}, \dots$.  This can be generally formulated

$$
{2m \choose m} \text{ for } m \in \mathbb{N} \land m \geq 1
$$

Using Pascal's identity, we can rewrite this to be
$$
{2m \choose m} = {2m-1 \choose m-1} + {2m-1 \choose m}, ~~ m \geq 1
$$

Because there is symmetry in Pascal's triangle row-wise e.g. ${n \choose k} = {n \choose n-k}$, then 
$$
{2m-1 \choose m-1} = {2m-1 \choose 2m-1-m+1} = {2m-1 \choose m}
$$

Therefore, 
$$
{2m \choose m} = {2m-1 \choose m} +{2m-1 \choose m} = 2{2m-1 \choose m}=2n, ~~ m \geq 1, n \in \mathbb{N}
$$

Since ${2m \choose m} =2n$ where n is a natural number then it must be the case that is an even number.

### 16.11
Use the Binomial Theorem to prove:
$$

$$

#### Answer
$$
(x+y)^n = \sum_{k=0}^n {n \choose k} x^k y^{n-k}
$$

If we let $x=-1$, and $y=1$, then
$$
0 = \sum_{k=0}^n {n \choose k}(-1)^k
$$
This is equilvant to our proposition. 

From a more combinatorial perspective, imagine we have a weirdo element that is $x \in U_n$ where $U_n = \{1, \dots, n \}$.  We denote $E_n$ to be the number of even sized subsets of $U_n$ and $O_n$ to be the number of odd sized subsets of $U_n$.  

$E_n$ can be constructed by looking at all even sized subsets of $U_n \setminus \{x\}$.  We will denote $F_{n-1}$ to be the number of such subsets.  This is only half the picture.  To fully construct $E_n$, we need to look at odd sized subsets of $U_n \setminus \{x\}$ and add $x$, the weirdo element, to make it even.  The number of such subsets is $G_{n-1}$

$$
E_n = F_{n-1} + G_{n-1}
$$

$O_n$ can be constructed by looking at all odd sized subsets of $U_n \setminus \{x\}$ (these subsets do not contain the weirdo element).  The number of such subsets is, as we noted above, $G_{n-1}$. The fact that we are adding $\{x\}$ to these subsets does not change the total number of such subsets, which as noted above, is denoted by $F_{n-1}$. 

$$
O_n = G_{n-1} + F_{n-1}
$$

Therefore, $E_n = O_n$.

$$

 \frac{4^{n} \sqrt{ \pi n}}{  \pi n }
$$
Since $ \pi n \geq \sqrt{\pi n} \Rightarrow 0 \leq \frac{ \sqrt{ \pi n}}{ \pi n } \leq 1$ and 

$$
{2n \choose n} \approx  \frac{4^{n} \sqrt{ \pi n}}{  \pi n } \leq 4^n
$$

Without using Stirling's formula, we will prove the inequality.  We will only use the fact that the cardinality of a power set of a set that contains n elements is $2^n$ and that a power set is a set of all possible subsets.  This can be represented using the below formula (from problem *16.10*).

### 16.20
Give a geometric and an algebraic proof that
$$
1 + 2 + 3 + \dots + (n-1) + n + (n-1) + (n-2) + \dots + 2 + 1 = n^2
$$

#### Answer 
* *Algebraic* proof goes as follows.  Because $ {n \choose 2} = 1 + 2 + 3 + \dots + (n-1)$, we have that

$$
{n \choose 2} + n + {n \choose 2} = 2\frac{n!}{2! (n-2)!} + n = \frac{n!}{(n-2)!} + n = n(n-1) + n = n^2 - n + n = n^2
$$

* *Geometric* proof goes as follows.  We count up the number of times a pair of two numbers sum to n and multiply by their sum i.e. the value $n$.  $1$ goes with $n-1$, $2$ goes with $n-2$ and so on.  This occurs $\frac{n-1}{2}$ times in the first half so we have: 
$$
n \Big( \frac{n-1}{2} \Big)
$$

This happens twice so this quantity becomes 
$$
2n \Big( \frac{n-1}{2} \Big) = n(n-1)
$$

We also need to add the middle term which is $n$, to get: 

$$
2n \Big( \frac{n-1}{2} \Big) = n(n-1) + n = n^2 - n + n  = n^2  
$$

Let A be an $n$-element set and suppose we have a supply of labels; we have $k$ labels that say "good" and $n-k$ labels that say "bad." #### Answer 
The answer is simply $ n \choose k$.

### 16.25
Let $n, a, b, c \in \mathbb{N}$ with $a+ b + c = n$. Please prove the following: 
* $\binom{n}{a, b, c} = \binom{n}{a}{n-a\choose b}$
#### Answer 
There are $n \choose a$ ways to label the first $a$ objects with Type 1. $n-a$ objects remain and of these, $b$ number of them need to be labeled with Type 2, so we have $n-a\choose b$.  Now, $n-a-b$ objects remain and of these, $c$ number of them need to be labeled Type 3, so we have ${n-a-b\choose c} = {c \choose c} = 1$.  By the product principle, the total number of way of labeling these objects is 
$$
\binom{n}{a, b, c}  = {n \choose a}{n-a\choose b}{n-a-b\choose c} = {n \choose a}{n-a\choose b}
$$
---

* $ \binom{n}{a, b, c} = \frac{n!}{a!b!c!}$
#### Answer 
Taking the result from the above problem, we have 
$$
\binom{n}{a, b, c} = {n \choose a}{n-a\choose b} = \frac{n!}{(n-a)!a!} \frac{(n-a)!}{b! (n-a-b)!} = \frac{n!}{a!b!(n-a-b)!}
$$
Since $ a + b + c = n$, then $ c = n - a - b$, so we get:

$$
\binom{n}{a, b, c} = \frac{n!}{a!b!c!}
$$

---

* If $a + b + c \ne n$, then $\binom{n}{a, b, c} = 0$
#### Answer 
Lets assume for the sake of contradiction that $\binom{n}{a, b, c} \ne 0$ and we have that
$$
\binom{n}{a, b, c} = \frac{n!}{(n-a)!a!} \frac{(n-a)!}{b! (n-a-b)!} = \frac{n!}{a! b! (n-a-b)!}
$$
Because $a + b + c \ne n$, then $c \ne n- a - b$ and wither $c > n- a - b$ or $c < n- a - b$.  If $c > n- a - b$ then we are asking to label $c$ items to be Type 3 when there are fewer items than $c$ available. It must be case that $\binom{n}{a, b, c} = 0$ hence a contradiction.  If $c < n - a -b$ then we are asking to label $c$ elements to be Type 3.  We can do this, but then there would be $n-a-b-c$ elements remaining that must be labed Type 1, Type 2or Type 3.  If we label them any one of these types then we contradict the fact that we are labeling $a$ to be Type 1, $b$ to be Type 2, and $c$ number of them to be Type 3, hence we reach a contradiction.  Therefore, it must be that if $a + b + c \ne n$, then $\binom{n}{a, b, c} = 0$.

Both quantities are equivalent and look like: 

$$
{13 \choose 1}{48 \choose 1} = {13 \choose 1}{12 \choose 1}{4 \choose 1} = 624
